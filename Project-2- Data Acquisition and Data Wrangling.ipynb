### Outliers

```python

q1=final_data['cnt'].quantile(.25)
q3=final_data['cnt'].quantile(.75)
    
iqr=q3-q1
min_r= q1- (1.5*iqr)
max_r= q3+ (1.5*iqr)
outliers=final_data.index[(final_data['cnt']<min_r) |  (final_data['cnt']>max_r)]  
```
